```python
import numpy as np
import os
import cv2
```

```python
one = os.listdir('./1/')
two = os.listdir('./2/')
three = os.listdir('./3/')
```

```python
x =[]
y = []
```

```python
for i in one:
    img = cv2.imread('./1/'+i, 0)
    img = cv2.resize(img, (32,32))
    x.append(img)
    y.append(1)
    print(i)
```

```python
for i in two:
    img = cv2.imread('./2/'+i, 0)
    img = cv2.resize(img, (32,32))
    x.append(img)
    y.append(2)
    print(i)
```

```python
for i in three:
    img = cv2.imread('./3/'+i, 0)
    img = cv2.resize(img, (32,32))
    x.append(img)
    y.append(3)
    print(i)
```
